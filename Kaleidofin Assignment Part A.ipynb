```python
 #Total Contribution Calculated by each person in each month is given by following formulae.
```

```python
CF_Data["Total Contribution"] = CF_Data["Contribution"] - CF_Data["Amount returned to everyone in the group"]
```

```python
CF_Data
```

```python
#Calculating Total Contribution contributed by each member of the chit fund throughout 25 months.
```

```python
Total = CF_Data["Total Contribution"].sum()
```

```python
Total
```

```python
#Calculating Annualized Return obtained by the Bid Winners every month. 
```

```python
CF_Data["Annualized Return"] = CF_Data["Net amount recd by Bid winner"] - Total
```

```python
CF_Data
```

```python
#Calculating Return % obtained by the Bid Winners every month
```

```python
CF_Data["Return %"] = (CF_Data["Annualized Return"]/Total)*100
```

```python
CF_Data
```
